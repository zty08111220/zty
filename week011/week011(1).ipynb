```python
import requests
import json

# 发送HTTP GET请求获取数据
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)

# 检查响应状态码，200表示成功
if response.status_code == 200:
    # 获取数据（假设数据是JSON格式）
    data = response.json()

    # 指定要保存的JSON文件路径和文件名
    json_file_path = "data.json"

    # 将数据写入JSON文件
    with open(json_file_path, 'w', encoding='utf-8') as json_file:
        json.dump(data, json_file, ensure_ascii=False, indent=4)

    print(f"数据已保存到 {json_file_path}")
else:
    print("请求失败")
```

```python
# 样例2 成功爬取

import requests
from bs4 import BeautifulSoup
import csv

# 设置豆瓣电影页面的URL

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
url = 'https://movie.douban.com/subject/1292722/reviews'

# 发送HTTP请求获取页面内容
response=requests.get(url,headers=headers)
if response.status_code == 200:
    soup = BeautifulSoup(response.text, 'html.parser')
else:
    print("无法访问网页")
    exit()

data = []

# 查找评论元素
reviews = soup.find_all('div', class_='review-item')
#print(reviews)
# 遍历评论元素并提取信息
for review in reviews:
    print(review)
    print("------------------------------------")
    reviewer = review.find('a', class_='name').text.strip()
    rating = review.find('span', class_='main-title-rating')
    if rating:
        rating = rating.text.strip()
    else:
        rating = '未评分'
    content = review.find('div', class_='short-content').text.strip()

    # 将提取的信息添加到数据列表中
    data.append([reviewer, rating, content])

# 将数据保存到CSV文件
with open('titanic_reviews.csv', 'w', newline='', encoding='utf-8') as csv_file:
    csv_writer = csv.writer(csv_file)
    csv_writer.writerow(['评论者', '评分', '评论内容'])  # 写入CSV文件头
    csv_writer.writerows(data)  # 写入数据行

print("数据已成功抓取并保存为CSV文件：titanic_reviews.csv")
```

```python

import requests
import csv

# 发送HTTP GET请求获取数据
url = "https://jsonplaceholder.typicode.com/posts"
response = requests.get(url)

# 检查响应状态码，200表示成功
if response.status_code == 200:
    # 获取数据（假设数据是列表或字典）
    data = response.json()

    # 指定要保存的CSV文件路径和文件名
    csv_file_path = "data.csv"

    # 将数据写入CSV文件
    with open(csv_file_path, 'w', newline='', encoding='utf-8') as csv_file:
        csv_writer = csv.writer(csv_file)

        # 如果数据是列表，可以直接写入
        if isinstance(data, list):
            csv_writer.writerows(data)
        # 如果数据是字典，可以从字典中提取数据并写入
        elif isinstance(data, dict):
            for key, value in data.items():
                csv_writer.writerow([key, value])

    print(f"数据已保存到 {csv_file_path}")
else:
    print("请求失败")
```

```python
#练习
import requests
import csv
from datetime import datetime
import re
url = "https://news.sina.com.cn/world"
response = requests.get(url)
# 获取网页内容
html_content = response.text.encode(response.encoding).decode() 
# 使用Beautiful Soup解析HTML
soup = BeautifulSoup(html_content, 'html.parser',from_encoding='utf-8')
news_list = soup.find_all('div',class_="news-item")

#正则化筛除
url_pattern = re.compile(r'https://news\.sina\.com\.cn/w/\d{4}-\d{2}-\d{2}/doc-\w+\.shtml')
#news_date = datetime.strptime(date_str, '%Y年%m月%d日 %H:%M')

current_date = datetime.now().strftime('%Y-%m-%d')
print(current_date)
# 创建一个 CSV 文件来存放数据
csv_filename = 'sina_news.csv'

with open(csv_filename, 'w', newline='', encoding='utf-8') as csv_file:
    # 创建 CSV 写入对象
    csv_writer = csv.writer(csv_file)
    # 写入表头
    csv_writer.writerow(['主题', '时间', '摘要'])
    for news_item in news_list:
        # 获取新闻链接
        news_link = news_item.find('a')
        if news_link is None:
            continue
        news_link = news_link.get('href')
        match = url_pattern.search(news_link)
        if match is None:
            continue
        
        #print(news_item)

        # 发起对新闻链接的请求
        news_response = requests.get(news_link)
        news_html_content = news_response.text.encode(news_response.encoding).decode()
        news_soup = BeautifulSoup(news_html_content, 'html.parser')
        
        
        # 获取新闻主题、时间和摘要
        #title = news_soup.title.text.strip()
        title = news_soup.find('h1',class_ = 'main-title').text
        print("title",title)
        #time_tag = news_soup.find('h1', cla ='weibo:article:create_at')
        time_tag = news_soup.find('span', class_ ="date")
        time = time_tag.text.strip() if time_tag else 'N/A'
        #逻辑：判断是否为最新一天的新闻
        news_date = datetime.strptime(time, '%Y年%m月%d日 %H:%M').strftime('%Y-%m-%d')
        if news_date == current_date:
            pass
        else:
            continue
        print("time",time)
        summary_tag = news_soup.find('meta', {'name': 'description'})
        summary = summary_tag['content'].strip() 
        print("summary",summary)

        # 写入 CSV 文件
        csv_writer.writerow([title, news_date, summary])

# 打印成功信息
print(f'新闻数据已成功抓取并保存到 {csv_filename}')
```
